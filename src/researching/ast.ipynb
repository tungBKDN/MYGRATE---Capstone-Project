```python
import ast

# 1. Create snippet (with Type Hints as in your problem)
source_code = """
const a = 3
const b = 2
let c = a + b
"""

# 2. Use the ast library to "parse" the text into a Syntax Tree
tree = ast.parse(source_code)

# 3. Dump (print) the tree for visualization (Requires Python 3.9+ for indent parameter)
print("=== AST TREE STRUCTURE ===")
print(ast.dump(tree, indent=4))
```
