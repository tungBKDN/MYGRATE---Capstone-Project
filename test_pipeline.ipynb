# MYGRATE — End-to-End Pipeline Test

Chạy toàn bộ pipeline qua orchestrator.  
Nhập input → orchestrator tự điều phối reader → architect → end.

```python
import json, sys, os
sys.path.insert(0, os.path.abspath('..') if os.path.exists('..') else '.')
from dotenv import load_dotenv
load_dotenv()

from src.workflow import build_app

auto_app = build_app(interrupt=False)
print("Workflow ready.")
```

```python
project_path = input("Project path: ")
target_java  = input("Target Java version (default 17): ") or "17"

initial_state = {
    "project_path": project_path,
    "target_java_version": target_java,
    "project_type": None,
    "source_framework": None,
    "source_version": None,
    "target_framework": None,
    "target_version": None,
    "messages": [],
    "completed_tasks_summary": [],
    "pom_data": None,
    "dependencies": [],
    "index_report": None,
    "upgrade_report": None,
    "candidate_solutions": None,
    "compatibility_matrix": {},
    "migration_tasks": [],
    "current_instruction": "",
    "last_subagent_result": "",
    "next_node": "supervisor",
}
```

```python
for event in auto_app.stream(initial_state):
    node_name = list(event.keys())[0]
    node_output = event[node_name]
    print(f"\n[{node_name}]")
    print(json.dumps(node_output, default=str, ensure_ascii=False, indent=2)[:2000])
```

```python
final_state = auto_app.invoke(initial_state)
print(json.dumps(final_state, default=str, ensure_ascii=False, indent=2)[:10000])
```
